# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = '5. Four By Two Productions'
```

```python
studios_list = pd.DataFrame(data['production_companies'].str.split('|').tolist()).stack().unique()
data['len_title'] = data.original_title.str.len()
len_list = {}
for studio in studios_list:
    f = data['len_title'][data.production_companies.map(lambda x: True if studio in x else False)].mean()
    len_list[studio] = f
title_len = pd.Series(len_list)
title_len.sort_values(ascending = False)
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['25'] = '3. Midnight Picture Show'
```

```python
studios_list = pd.DataFrame(data['production_companies'].str.split('|').tolist()).stack().unique()
s = int(data['overview'].count())
length=[]
for i in range(0,s):
    l = len(data.overview.str.split(' ')[i])
    length.append(l)
data['len_overview'] = length
    
len_owerviews = {}
for studio in studios_list:
    f = data['len_overview'][data.production_companies.map(lambda x: True if studio in x else False)].mean()
    len_owerviews[studio] = f
owerview = pd.Series(len_owerviews)
owerview.sort_values(ascending = False) 
```
